```python
# Cell 2: Upload this zip and extract it
# (Run this cell, then use the file browser to upload stealthwatch_training.zip)
from google.colab import files
import zipfile, os

uploaded = files.upload()   # upload stealthwatch_training.zip here
zip_name = list(uploaded.keys())[0]
with zipfile.ZipFile(zip_name, 'r') as z:
    z.extractall('.')
os.chdir('stealthwatch_training')
print('Extracted. Current dir:', os.getcwd())
print('Files:', os.listdir('.'))
```

```python
# Cell 3: Upload CTU-13 .labeled files
# Download from: https://www.stratosphereips.org/datasets-ctu13
# Recommended scenarios: 1, 2, 9, 13
import os
os.makedirs('data/raw/CTU-13', exist_ok=True)

from google.colab import files
print('Upload your .labeled files now (can select multiple):')
uploaded = files.upload()
for fname in uploaded:
    dest = f'data/raw/CTU-13/{fname}'
    with open(dest, 'wb') as f:
        f.write(uploaded[fname])
    print(f'  Saved → {dest}')

import glob
found = glob.glob('data/raw/CTU-13/*.labeled')
print(f'\nFound {len(found)} .labeled file(s):', found)
```
